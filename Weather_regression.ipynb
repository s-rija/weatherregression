```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/dataset/weatherHistory.csv')
df.head()
```

```python
df2=df[['Temperature (C)','Humidity','Precip Type']].copy()
df2.head()
#'Precip Type','Apparent Temperature (C)','Wind Speed (km/h)'
```

Finding the unique types of feature.

```python
#print(df.Summary.unique())
print(df['Precip Type'].unique())
#print(df['Daily Summary'].unique())
df2['Precip Type']=df2['Precip Type'].fillna(0)# To fill the null(nan) with 0.
print(df2['Precip Type'].unique())
df2['Precip Type'].replace(['rain', 'snow'],[0, 1], inplace=True)# To replacge the feature types with numerical values.
df2.head()
print(df2['Precip Type'].unique())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df2.drop(['Humidity'], axis = 1), df2['Humidity'], test_size=0.25)# Temp and Preciptype calculating to find the humidity(output).
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(x_train, y_train)
print(regr.score(x_test, y_test))
```
